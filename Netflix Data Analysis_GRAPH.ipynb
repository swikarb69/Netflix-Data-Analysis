# 1. Find the most common genres.

```python
netflix_titles
```

```python
netflix_titles['listed_in'] = netflix_titles['listed_in'].str.split(",")
```

```python
genre_df = netflix_titles.explode('listed_in')
```

```python
genre_df['listed_in'] = genre_df['listed_in'].str.strip()
```

```python
genre_df['listed_in'].value_counts()
```

```python
genre_counts = genre_df['listed_in'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Number of movies']
top_genres = genre_counts.head(15)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Number of movies', y='Genre', data=top_genres)
plt.title('Top 15 Most Common Genres')
plt.show()
```

## Insight: Top 15 Most Common Genres on Netflix

Netflix's genre distribution reveals a strong preference for **drama, international, and comedy content**, with a long tail of more niche categories.

- **International Movies top the chart** with ~2,750 titles, making it the single largest genre tag — highlighting Netflix's global content strategy
- **Dramas (~2,400)** and **Comedies (~1,650)** are the dominant traditional genres, suggesting these are the core of Netflix's appeal
- **International TV Shows (~1,350)** rank 4th, confirming that non-English series (K-dramas, anime, Spanish telenovelas, etc.) are a major part of the catalog
- **Documentaries and Action & Adventure** are nearly tied at ~850 each, showing strong demand for both factual and high-energy content
- The bottom half of the chart (TV Comedies, Thrillers, Crime TV Shows, Kids' TV, Docuseries) all cluster between **400–600 titles**, indicating a fairly balanced long tail

**Takeaway:** Netflix clearly bets big on international and dramatic content. The prominence of "International Movies" as the #1 genre tag reflects the platform's deliberate push beyond Hollywood to serve a worldwide audience.

# 2. Identify which country produces the most content.

```python
netflix_titles['country'] = netflix_titles['country'].str.split(",")
```

```python
country_df = netflix_titles.explode('country')
```

```python
country_df['country'] = country_df['country'].str.strip()
```

```python
country_df['country'].value_counts()
```

```python
country_counts = country_df['country'].value_counts().reset_index()
country_counts.columns = ['Countries', 'Movies Produced']
top_countries = country_counts.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Movies Produced', y='Countries', data = top_countries)
plt.title('Top 5 Countries with highest number of movies produced')
plt.show()
```

## Insight: Top 5 Countries with Highest Number of Movies Produced

The United States dominates Netflix's content library by an enormous margin, with the next closest country producing less than a third as much content.

- **United States leads overwhelmingly** with ~3,600+ titles — more than all other top countries combined
- **India ranks 2nd** with ~1,050 titles, reflecting Netflix's heavy investment in Bollywood and regional Indian content
- **"Unknown"** appears as the 3rd largest category (~850 titles), meaning a significant portion of the library has no country metadata — this is a data quality issue worth noting
- **United Kingdom (~800)** and **Canada (~450)** round out the top 5, both English-speaking markets closely tied to US production pipelines

**Takeaway:** Netflix's library is heavily US-centric, but India's strong presence signals a clear push toward global and regional content diversification.

# 3. Analyze trends of releases by year.

```python
netflix_titles
```

```python
netflix_titles['release_year'].value_counts()
```

```python
years_counts = netflix_titles['release_year'].value_counts().reset_index()
years_counts.columns = ['Years', 'Number of Movie released']
top_trends = years_counts.head(10)
```

```python
plt.figure(figsize = (12, 8))
sns.barplot(x = 'Years', y = 'Number of Movie released', data = top_trends)
plt.title('Top 10 trending movies by Years')
plt.xticks(rotation=45)
plt.show()
```

## Insight: Top Trending Years for Netflix Content Releases

Release volume shows a **clear upward trend through the 2010s**, reflecting Netflix's massive content expansion strategy during that decade.

- **2018 was the peak year**, with ~1,150 titles released — Netflix's most prolific year on record in this dataset
- Growth accelerated sharply from **2015 onwards**, nearly doubling from ~560 (2015) to over 1,000 (2017)
- **2021 shows a drop to ~600**, which again is likely a **data cutoff artifact** rather than a real slowdown
- The x-axis labels show only the last 2 digits of the year (12–21), representing **2012 to 2021**

**Takeaway:** Netflix was in full content-production overdrive between 2016–2020, adding hundreds of new titles every year as it competed aggressively in the global streaming wars.

# 4. Compare TV Shows vs Movies.

```python
netflix_titles
```

```python
netflix_titles['type'].value_counts()
```

```python
type_year = netflix_titles.groupby(['release_year', 'type']).size().unstack(fill_value = 0)
type_year
```

```python
fig, ax = plt.subplots(figsize=(14, 6))
type_year.plot(ax=ax)
ax.set_title('Movies vs TV Shows Released Over the Years')
ax.set_xlabel('Release Year')
ax.set_ylabel('Number of Releases')
plt.show()
```

## Insight: Movies vs TV Shows Released Over the Years

Both content types remained **nearly flat until the 2000s**, after which Netflix's original content push drove an explosive rise in releases.

- **Movies consistently outnumber TV Shows** across all years, peaking at ~770 releases around **2019–2020**
- **TV Shows followed the same growth curve** but at a lower volume, peaking at ~430 around the same period
- The **sharp drop after 2020** is likely a **data cutoff issue** — the dataset doesn't fully capture recent years, not an actual decline in production
- Growth was essentially **flat pre-2000**, meaning the vast majority of Netflix's library is recent content

**Takeaway:** Netflix aggressively expanded its content library through the late 2010s, with movies leading the charge. The post-2020 dip is a dataset artifact, not a real trend.

# 5. Find the longest movie duration.

```python
netflix_titles
```

```python
netflix_titles['duration'].dtype
```

```python
netflix_titles['duration'].value_counts()
```

```python
movies_df = netflix_titles[netflix_titles['type'] == 'Movie'].copy()
```

```python
movies_df['duration'].unique()
```

```python
movies_df = movies_df[movies_df['duration'] != 'Unknown']
```

```python
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(int)
```

```python
movies_df['duration'].max()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(movies_df['duration'], bins=30)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

## Insight: Distribution of Movie Durations

The distribution is **right-skewed**, with most Netflix movies falling between **80–110 minutes** — the sweet spot for a typical feature film.

- The **peak** is around 90–100 minutes, with over 1,200 movies in that range
- Very few movies exceed **200 minutes**, indicating Netflix rarely hosts extremely long films
- The tail stretches to ~312 minutes, which represents outliers (documentaries, special cuts, etc.)

**Takeaway:** Netflix's movie library is heavily optimized for standard feature-length content (~1.5 hrs), likely to suit average viewer attention spans.
